```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('fin_int.csv')
data
```

```python
data = data[data['Area(ha)'] != 0]
data = data[data['Yield(kg/ha)'] != 0]
data
```

```python
data = data.drop('District', axis=1)
data
```

```python
data = data.drop('Production(mt)', axis=1)
data = data.drop('Area(ha)', axis=1)
data
```

# First

```python
from sklearn.preprocessing import StandardScaler

X = data.drop('Yield(kg/ha)', axis=1)  # Input features
y = data['Yield(kg/ha)']  # Target variable

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

display(X_scaled_df)
display(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)
display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

feature_importances = rf.feature_importances_
importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})
importance_df = importance_df.sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(importance_df['Feature'], importance_df['Importance'])
plt.xticks(rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, explained_variance_score
import numpy as np

y_pred = rf.predict(X_test)

r2 = r2_score(y_test, y_pred)
print("R-squared score:", r2)

n = X_test.shape[0]  # Number of samples in the test set
p = X_test.shape[1]  # Number of features
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
print("Adjusted R-squared score:", adjusted_r2)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)

explained_var = explained_variance_score(y_test, y_pred)
print("Explained Variance Score:", explained_var)

mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print("Mean Absolute Percentage Error:", mape)
```

# Second

```python
selected_features = ['Year', 'Latitude', 'Longitude', 'Precipitation(mm)', 'Wind_Speed_50M(m/s)', 'Relative_Humidity(%)', 'Surface_Pressure(kPa)', 'Specific_Humidity(g/kg)', 'Crop_Barley', 'Crop_Maize', 'Crop_Millet', 'Crop_Paddy', 'Crop_Wheat', 'Yield(kg/ha)']
data_selected = data[selected_features].copy()
data_selected
```

```python
X_selected = data_selected.drop('Yield(kg/ha)', axis=1)
y = data['Yield(kg/ha)']
display(X_selected)
display(y)
```

```python
scaler = StandardScaler()
X_scaled_selected = scaler.fit_transform(X_selected)
X_scaled_selected_df = pd.DataFrame(X_scaled_selected, columns=X_selected.columns)

X_scaled_selected_df
```

```python
# Split the scaled selected features and the target variable into training and testing sets
X_train_selected, X_test_selected, y_train, y_test = train_test_split(X_scaled_selected_df, y, test_size=0.2, random_state=42)

# Train the Random Forest model on the training data
rf.fit(X_train_selected, y_train)

# Make predictions on the test set
y_pred_selected = rf.predict(X_test_selected)

# Calculate and print the evaluation metrics
r2_selected = r2_score(y_test, y_pred_selected)
adjusted_r2_selected = 1 - ((1 - r2_selected) * (n - 1) / (n - p - 1))
mae_selected = mean_absolute_error(y_test, y_pred_selected)
mse_selected = mean_squared_error(y_test, y_pred_selected)
rmse_selected = np.sqrt(mse_selected)
explained_var_selected = explained_variance_score(y_test, y_pred_selected)
mape_selected = np.mean(np.abs((y_test - y_pred_selected) / y_test)) * 100

print("R-squared score (selected features):", r2_selected)
print("Adjusted R-squared score (selected features):", adjusted_r2_selected)
print("Mean Absolute Error (selected features):", mae_selected)
print("Mean Squared Error (selected features):", mse_selected)
print("Root Mean Squared Error (selected features):", rmse_selected)
print("Explained Variance Score (selected features):", explained_var_selected)
print("Mean Absolute Percentage Error (selected features):", mape_selected)
```

```python
importance_selected = rf.feature_importances_
importance_df_selected = pd.DataFrame({'Feature': X_selected.columns, 'Importance': importance_selected})
importance_df_selected = importance_df_selected.sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(importance_df_selected['Feature'], importance_df_selected['Importance'])
plt.xticks(rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances (Selected Features)')
plt.show()
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
    # Add more hyperparameters if desired
}

rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, verbose=2, scoring='neg_mean_squared_error')

grid_search.fit(X_train_selected, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", np.sqrt(-grid_search.best_score_))
```

```python
best_rf = grid_search.best_estimator_
y_pred_tuned = best_rf.predict(X_test_selected)
y_pred_tuned
```

```python
r2_tuned = r2_score(y_test, y_pred_tuned)
adjusted_r2_tuned = 1 - ((1 - r2_tuned) * (n - 1) / (n - p - 1))
mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
mse_tuned = mean_squared_error(y_test, y_pred_tuned)
rmse_tuned = np.sqrt(mse_tuned)
explained_var_tuned = explained_variance_score(y_test, y_pred_tuned)
mape_tuned = np.mean(np.abs((y_test - y_pred_tuned) / y_test)) * 100

print("R-squared score (tuned model):", r2_tuned)
print("Adjusted R-squared score (tuned model):", adjusted_r2_tuned)
print("Mean Absolute Error (tuned model):", mae_tuned)
print("Mean Squared Error (tuned model):", mse_tuned)
print("Root Mean Squared Error (tuned model):", rmse_tuned)
print("Explained Variance Score (tuned model):", explained_var_tuned)
print("Mean Absolute Percentage Error (tuned model):", mape_tuned)
```
