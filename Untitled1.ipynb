```python
import pandas as pd

# Load the dataset
data = pd.read_csv('encoded_file_district.csv')
data
```

```python
data.drop('District', axis=1, inplace=True)
data
```

```python
data = data[data['Area(ha)'] != 0]
data = data[data['Yield(kg/ha)'] != 0]
data
```

```python
data = data[['Year', 'Latitude', 'Longitude', 'Precipitation(mm)', 'Temp(C)', 'Area(ha)', 'Production(mt)', 'Crop_Barley', 'Crop_Maize', 'Crop_Millet', 'Crop_Paddy', 'Crop_Wheat']]
data
```

```python
# data.to_csv('final_dataset.csv', sep=',', index=False, encoding='utf-8')
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target variable
features = data.drop(['Production(mt)'], axis=1)  # Drop the target variable
target = data['Production(mt)']

columns_to_scale = ['Year', 'Latitude', 'Longitude', 'Precipitation(mm)', 'Temp(C)', 'Area(ha)']

# Scale the features using StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features[columns_to_scale].values)
features[columns_to_scale] = scaler.transform(features[columns_to_scale].values)

features
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor

# Train a GradientBoostingRegressor model
gb_regressor = GradientBoostingRegressor()
gb_regressor.fit(X_train.values, y_train.values)

# Predict on the test set
y_pred = gb_regressor.predict(X_test.values)

# Predict on new inputs
inputs = [[1982, 29.11, 81.3, 1.822329, 19.098575, 500.0]]  # Example new inputs, adjust as needed
scaled_inputs = scaler.transform(inputs)
final_inputs = np.concatenate((scaled_inputs.reshape(6,), [1, 0, 0, 0, 0])).reshape(1, -1)
display(final_inputs)
output = gb_regressor.predict(final_inputs)

print('Predicted production for new inputs:', output)
```

```python
negative_indices = [value for index, value in enumerate(y_pred) if value < 0]

if negative_indices:
    print("Negative values found in y_pred:", negative_indices)
else:
    print("No negative values found in y_pred")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train.values, y_train.values)

# Predict on the test set
y_pred = rf_regressor.predict(X_test.values)

# Predict on new inputs
inputs = [[1982, 29.11, 81.3, 1.822329, 19.098575, 500.0]]  # Example new inputs, adjust as needed
scaled_inputs = scaler.transform(inputs)
final_inputs = np.concatenate((scaled_inputs.reshape(6,), [1, 0, 0, 0, 0])).reshape(1, -1)
display(final_inputs)
output = rf_regressor.predict(final_inputs)

print('Predicted production for new inputs:', output)
```

```python
import joblib

# Save the trained model to a file
joblib.dump(rf_regressor, 'rf_regressor.pkl')
```
