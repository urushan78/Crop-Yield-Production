We can see that the above dataset is not consistent with the climate dataset. So, some processing needs to be done before combining this dataset with the climate dataset.<br>
Lets make a funtion for that below.

```python
def process_missing_dataset(df, climate_df):
    df.rename(columns = {'MO':'MONTH', 'DY':'DAY', 'PRECTOTCORR':'PRECTOT'}, inplace=True) #rename columns 'MO' to 'MONTH' and 'DY' to 'DAY' and 'PRECTOTCORR to 'PRECTOT'
    df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH', 'DAY']]) #combine the 'YEAR', 'MONTH', 'DAY' columns into a single 'DATE' column
    df.drop(columns=['YEAR', 'MONTH', 'DAY'], inplace=True) #drop the 'YEAR', 'MONTH', 'DAY' columns
    df['T2M_RANGE'] = df['T2M_MAX'] - df['T2M_MIN'] #correct the temperature range column
    climate_df = pd.concat([climate_df, df], axis=0) #Combine the missing dataset with the climate dataset.
    return climate_df
```

Now, lets add all the missing data to the climate data.

```python
for missing_district in missing_districts:
    missing_df = pd.read_csv(f'missing_districts/{missing_district.lower()}.csv')
    missing_df['DISTRICT'] = missing_district
    climate_df = process_missing_dataset(missing_df, climate_df)
climate_df.sort_values(by=['DISTRICT', 'DATE'], ascending=[True, True], inplace=True)
climate_df
```

Now, lets convert this daily data to fiscal yearly data because the crop data that we have is in fiscal year.<br>
A fiscal year in Nepal starts from July 16 and ends in July 15 of next year.

```python
climate_df['YEAR'] = climate_df.DATE.dt.year #make a new 'YEAR' column and set the value to the corresponding year from the 'DATE' column
climate_df.loc[(climate_df.DATE.dt.month > 7) |
               ((climate_df.DATE.dt.month == 7) &
                (climate_df.DATE.dt.day > 15)), 'YEAR'] += 1 #if date is above July 15, increase the value in year column by 1.
climate_df = climate_df[climate_df['DATE'] >= pd.to_datetime('1982-07-16')] #drop all data before 1982-07-16
climate_df = climate_df[climate_df['DATE'] <= pd.to_datetime('2019-07-15')] #drop all data after 2019-07-15
climate_df.YEAR = (climate_df.YEAR - 1).astype(str) + '/' + (climate_df.YEAR).astype(str) #modify the 'YEAR' column so that it represents fiscal year instead of regular year
climate_df.reset_index(drop=True, inplace=True) #reset the index numbering and start from 0
climate_df
```

```python
crop_df.drop(columns=['DISTRICT_CODE'], inplace=True) #drop the 'DISTRICT_CODE' column
crop_df = crop_df.rename(columns={'DISTRICT_NAME': 'Year'}).set_index('Year').T #transpose the dataset
crop_df
```

The crop name(paddy, wheat, maize, millet, barley), stat type(production, area, yield) and the fiscal year are all in the same 'YEAR' column. They need to be separeted. Let's make functions to separate them below.

```python
def get_crop(col):
    return {
        'PD': 'Paddy',
        'WT': 'Wheat',
        'MZ': 'Maize',
        'ML': 'Millet',
        'BL': 'Barley',
        'BW': 'Buckwheat'
    }[col.split('_')[0]]


def get_stats_type(col):
    return {
        'P': 'Production',
        'A': 'Area',
        'Y': 'Yield',
    }[col.split('_')[1]]


def get_year(col):
    year = col.split('_')[2][:4]
    return f'{year}/{int(year) + 1}'
```

```python
crop_df['Crop'] = crop_df.index.map(get_crop) #Add a 'Crop' column containing the corresponding crops
crop_df['Stat_Type'] = crop_df.index.map(get_stats_type) #Add a 'Stat_Type' column containing the corresponding stat type.
crop_df.index = crop_df.index.map(get_year) #Change the year column so that it contains only the fiscal year
crop_df
```

```python
crop_df.reset_index(inplace=True) #reset the index so that 'Year' column is no longer the index
crop_df.rename(columns={'index':'Year'}, inplace=True) #Rename the column 'index' to 'Year'
crop_df
```

```python
#Change all the columns except 'Year', 'Crop' and 'Stat_Type' into values of a single column named 'District' and
#all the corresponding values in a column named 'Value'
crop_df = crop_df.melt(id_vars=['Year', 'Crop', 'Stat_Type'],
                       var_name='District',
                       value_name='Value')
crop_df
```

```python
#Make each value of 'Stat_Type' into a separate column with 'Year', 'District' and 'Crop' as index and the values
#in the column 'Value' as values
crop_df = crop_df.pivot_table(index=['Year', 'District', 'Crop'], columns='Stat_Type', values='Value')
crop_df
```

```python
#remove the column 'Crop' from index.
crop_df = crop_df.reset_index('Crop')
crop_df
```

We need to convert the fiscal year in string into a numeric form. To do that, let's represent a fiscal year using the starting year of the fiscal year. For example, 1982/1983 will be represented as 1982. To do that, let's make a function below to extract the initial year from the string like 1982/1983

```python
def convert_year(col):
    year = col.split('/')[0]
    return year
```
