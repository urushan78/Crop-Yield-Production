```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import KFold

# Define the features and target variable
X = data.drop('Production(mt)', axis=1)
y = data['Production(mt)']

# Initialize the Random Forest model
random_forest = RandomForestRegressor()

# Define the evaluation metrics
metrics = {
    'R2 Score': make_scorer(r2_score),
    'Mean Absolute Error (MAE)': make_scorer(mean_absolute_error),
    'Mean Squared Error (MSE)': make_scorer(mean_squared_error),
    'Root Mean Squared Error (RMSE)': make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred))),
    'Explained Variance Score (EVS)': make_scorer(explained_variance_score),
    'Adjusted R2 Score': make_scorer(lambda y_true, y_pred: 1 - ((1 - r2_score(y_true, y_pred)) * (len(y_true) - 1)) / (len(y_true) - X.shape[1] - 1)),
    'Mean Absolute Percentage Error (MAPE)': make_scorer(mean_absolute_percentage_error)
}

# Perform cross-validation and calculate evaluation metrics
cv = KFold(n_splits=5, shuffle=True, random_state=42)

for metric_name, scorer in metrics.items():
    scores = cross_val_score(random_forest, X, y, cv=cv, scoring=scorer)
    mean_score = np.mean(scores)
    std_score = np.std(scores)
    print(f"{metric_name}: {mean_score:.4f} +/- {std_score:.4f}")
```
