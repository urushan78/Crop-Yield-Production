```python
import pandas as pd
import json

# Read the CSV file using pandas
df = pd.read_csv('he.csv')

# Calculate averages for each district
averages = df.groupby('District').mean().to_dict()

# Convert averages to JSON format
averages_json = json.dumps(averages)

# Save the averages as a JSON file
with open('averages.json', 'w') as jsonfile:
    jsonfile.write(averages_json)

print("Averages file created!")
```
